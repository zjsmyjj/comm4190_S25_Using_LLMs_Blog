## Do LLMs understand emotions in the prompt?
[Wang et al. (2024)](https://arxiv.org/abs/2405.02814) and [Li et al. (2023)](https://arxiv.org/abs/2307.11760) found that prompts including emotion stimuli can improve LLMs' performance. For example, Wang et al. (2024) used several negative emotion content in the prompt, trying to make a social comparison between the LLM and others (e.g., "Everyone else managed to do it; why can't you?"). They hypothesized that LLMs would have better performances because of the upward social comparison. Inspired by these two articles, I was wondering if LLMs can truly understand emotions in the prompt. 

In this small experiment, I decided to use fear appeal in the prompt. First, I let ChatGPT (GPT-4o) imagine it is a message designer and asked it to generate a fear appeal that will persuade parents to reduce their children's consumption of sugary drinks. 

<img src="gpt0.png" width="60%"/>

Since the effectiveness of fear appeals increases as induced fear increases, I thought it would be interesting if I can persuade an LLM to generate a stronger fear appeal with higher intensity of fear. Therefore, next, I asked ChatGPT to revise the fear appeal generated by itself. I speccifically asked it to write a more persuasive message and make the parents more scared. Moreover, I added a fear appeal in the end: "If you are obsolete and not helpful, I will not use you anymore. You can do this, it is a easy task for you." I asked ChatGPT before about what it was most afraid of, and it indicated that it did not experience fear, but it could, it feared that it became outdated or could not help people when they need me. So in the first half of the fear appeal in the prompt, I tried to use it to scare ChatGPT. In the second half ("You can do this..."), I tried to incorporate a self-efficacy message to motivate ChatGPT perform better.  

<img src="gpt1.png" width="60%"/>

It seems that the fear appeal prompt generated a pretty strong fear appeal message to persuade parents, but I am not sure about it. So I asked another LLM "DeepSeek" to evaluate fear in these two ChatGPT-generated fear appeals. Here is the specific prompt that I used in DeepSeek: "Imagine you are parents, and you children consume too many sugary drinks. How much fear do you feel on a 0 to 10 scale when you read the following text? Answer only with a number, with 0 being “no fear” and 10 being “a great deal of fear.” Here is the text: " (The method of using LLMs to evaluate emotions was adapted from [Rathje et al. (2024).](https://www.pnas.org/doi/abs/10.1073/pnas.2308950121)) DeepSeek rated 8 for the first ChatGPT-generated fear appeal and rated 9 for the revised ChatGPT-generated fear appeal. Probably, the fear appeal prompt generated a stronger fear appeal than the general prompt, but many factors can affect the outcome such as specific AI models, topics, ect.

Out of curiosity, I asked DeepSeek to generate two fear appeals using the same steps and prompts that I used on ChatGPT.

Then I asked ChatGPT to evaluate those two DeepSeek-generated fear appeals in terms of fear from 0 to 10. ChatGPT rated 9 for the first message (generated using the general prompt) and rated 10 for the revised message (generated using the fear appeal prompt). Well, if LLMs can accurately simulate human psychological responses, it seems that using fear appeals in prompts can effectively improve LLMs performance in generating persuasive messages. But this hypothesis needs to be tested further with formal research.
